# Create new dataframe with all the cancelled items

```python
cancelled_items=df.loc[df['Invoice'].str[0]=='C']
cancelled_items
```

```python
df.drop(df[df['Invoice'].str[0]=='C'].index, inplace = True)
#df.reset_index(inplace = True)
print(len(df))
df.head(190)
```

# Bargraph of quantity for each country for all cancelled items

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='Quantity', y = 'Country', ax=ax, data=cancelled_items)

#plt.figure(figsize=(30,4))
```

# Date time index seperation 

```python
date_time_df = df

date_time_df.head()
```

```python
date_time_df['new_date'] = [d.date() for d in date_time_df['InvoiceDate']]
date_time_df['new_time'] = [d.time() for d in date_time_df['InvoiceDate']]
```

```python
date_time_df.head(200)
```

```python
date_time_df_ascend = date_time_df.sort_values('new_date', ascending=True)
#fig_dims = (20, 10)
#fig, ax = plt.subplots(figsize=fig_dims)
plt.plot(date_time_df_ascend['new_date'], date_time_df_ascend['Quantity'])
plt.xticks(rotation='vertical')
```
